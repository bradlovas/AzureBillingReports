```python
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
import pathlib
```

```python
reports_dir = pathlib.Path().resolve().parent.parent.parent / "Desktop" / "Carrier Reports" / "Data"
reports_dir
```

```python
file_name = "usage-2021-11-16T14-19-08+00-00.csv-filtered.csv"
file_path = pathlib.Path(reports_dir).joinpath(file_name)
file_path
```

```python
csv_database = create_engine('sqlite:///csv_database.db')
```

```python
chunk_size=50000
batch_no=1
for chunk in pd.read_csv(file_path,chunksize=chunk_size,iterator=True):
    chunk.to_sql('chunk_sql',csv_database, if_exists='append')
    batch_no+=1
    print('index: {}'.format(batch_no))
```
